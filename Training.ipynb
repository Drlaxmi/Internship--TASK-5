```python
import pandas as pd
```

```python
dataset=pd.read_csv("train.csv")
```

```python
dataset
```

```python
#STATISTICAL EXPLORATION
```

```python
dataset.info()              # Structure of dataset
dataset.describe()          # Summary statistics
dataset.isnull().sum()      # Missing values
dataset['Survived'].value_counts()  # Target class distribution
```

```python
#VISUAL EXPLORATION
```

```python
#UNIVARIATE VARIABLES
import matplotlib.pyplot as plt
import seaborn as sns

dataset['Age'].hist()
sns.countplot(x='Survived', data=dataset)
```

```python
#BIVARIATE VARIABLES
sns.boxplot(x='Survived', y='Age', data=dataset)
sns.countplot(x='Sex', hue='Survived', data=dataset)
```

```python
#MULTIVARIATE VARIABLES
sns.heatmap(dataset.corr(numeric_only=True), annot=True)
sns.pairplot(dataset[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
```
